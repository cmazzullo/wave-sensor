```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import mpld3
import NetCDF_Utils.grapher as grapher
import pandas as pd
from DepthCalculation.depth import Depth as depth
from functools import reduce
import IPython.html.widgets as widgets
plt.rcParams['figure.figsize'] = (10, 5)


mpld3.enable_notebook()
```

```python

class PlotWave(object):
    
    def __init__(self):
        self.wave = np.linspace(-np.pi, np.pi, 200)
        self.amp = 1
        self.period = 1
        self.cos = False
        self.equation = '(%ssin(x*%s))' %(self.amp,self.period)
        
    def get_wave(self):
        if self.cos == False:
            self.equation = '(%ssin(x*%s))' %(self.amp,self.period)
            return np.multiply(self.amp,np.sin(self.wave*self.period))
        else:
            self.equation = '(%scos(x*%s))' %(self.amp,self.period)
            return np.multiply(self.amp,np.cos(self.wave*self.period))
        
```

```python
z_offset = [0]
values = []
pressure_data = [0]
for x in range(0,6):
    values.append(PlotWave())
    
def change_z(z):
    z_offset[0] = z
    f()
    
def f():
    x = np.linspace(-np.pi, np.pi, 200)
    pressure_data[0] = np.add(reduce(np.add,[x.get_wave() for x in values]),z_offset[0])
    plt.plot(x,np.add(reduce(np.add,[x.get_wave() for x in values]),z_offset[0]))
    plt.grid(b=True, which='major', color='grey', linestyle="-",alpha=0.5)
    plt.show()
    
def change_value(plot_num, amp, period, cos):
  
    plot = values[plot_num]
    if amp != None:
        plot.amp = amp
    if period != None:
        plot.period = period
    if cos != None:
        plot.cos = cos
        
    f()
```

```python
lims = (1,20,2)
alims = (1,5,1)
z_lims = (-5,15,1)
widgets.interact(change_z, z = z_lims)
for x in range(0,len(values)):
    widgets.interact(change_value, plot_num = widgets.fixed(x), amp = alims, period = lims, cos = False)
    
```
