```python

class PlotWave(object):
    
    def __init__(self):
        self.wave = np.linspace(-np.pi, np.pi, 200)
        self.amp = 1
        self.period = 1
        self.cos = False
        self.equation = '(%ssin(x*2pi*%s))' %(self.amp,self.period)
        
    def get_wave(self):
        if self.cos == False:
            self.equation = '(%ssin(x*2pi*%s))' %(self.amp,self.period)
            return np.multiply(self.amp,np.sin(self.wave *(2*np.pi) * self.period))
        else:
            self.equation = '(%scos(x*2pi*%s))' %(self.amp,self.period)
            return np.multiply(self.amp,np.cos(self.wave * (2*np.pi) * self.period))
        
z_offset = [0]
values = [PlotWave(),PlotWave(),PlotWave(),PlotWave(),PlotWave(),PlotWave()]
# value = [PlotWave() for x in range(0,6)]
pressure_data = [0]
```

```python
def change_z(z):
    z_offset[0] = z
    f()
    
def f():
    x = np.linspace(-np.pi, np.pi, 200)
    pressure_data[0] = np.add(reduce(np.add,[x.get_wave() for x in values]),z_offset[0])
    plt.plot(x,np.add(reduce(np.add,[x.get_wave() for x in values]),z_offset[0]))
    plt.grid(b=True, which='major', color='grey', linestyle="-",alpha=0.5)
    plt.show()
    
    
def change_value(plot_num, amp, period, cos):
  
    plot = values[plot_num]
    if amp != None:
        plot.amp = amp
    if period != None:
        plot.period = period
    if cos != None:
        plot.cos = cos
        
    f()
    
from NetCDF_Utils.ExportCombinationGraph import export_netCDF

def export(a):
    
    z_offset[0] = str(z_offset[0])
    equation = ''.join(['%s + ' % x.equation for x in values] + z_offset)
    print(equation)
    exp = export_netCDF(pressure_data[0],equation)
    exp.export()

def print_hello(a):
    print('hey')
```

```python
lims = (1,10,1)
alims = (1,5,1)
z_lims = (-5,15,1)
widgets.interact(change_z, z = z_lims)
for x in range(0,len(values)):
    widgets.interact(change_value, plot_num = widgets.fixed(x), amp = alims, period = lims, cos = False)
    
```

```python
import types
from functools import partial
import json

class JSCode(object):
    def __init__(self, code):
        self.code = code
        self.invokes = ""
        
        # Find all the function name(  lines in the Javasript code
        for line in code.split("\n"):
            line = line.strip().split()
            if line and line[0] == 'function':
                funcname = line[1].split("(")[0]
                
                # For each create a bound method to add to the list of invocations
                def invoke(self, *args):
                    argstr = ','.join([json.dumps(arg) for arg in args])
                    self.invokes += "%s(%s);\n" % (funcname, argstr)                

                setattr(self, funcname, types.MethodType(invoke, self))    
                
    def _repr_javascript_(self):
        """Return the code definitions and all invocations"""
        return self.code + self.invokes
```
