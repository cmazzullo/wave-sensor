```python
Q: What are the goals of this notebook?
A: I want to create a file with known wavelengths, etc, in waveheight (meters)
I want to convert that to pressure using LWT
Then I want to convert that pressure back to wave height using our algorithm
And I want to automate this whole process.
```

```python
from numpy import *
from matplotlib.pyplot import *
%matplotlib inline
from scipy.optimize import newton
from pressure_to_depth import fft_method

g = 9.8
h = 20
rho = 1027
z = -10
```

```python
def make_waves(length, sample_frequency, waves):
    """Create wave pressure given frequencies, amplitudes and phases"""
    t = arange(length, step=1/sample_frequency)
    total_height = zeros_like(t)
    total_pressure = zeros_like(t)
    for wave in waves:
        f, a, phi = wave
        eta = a*sin(2*pi*f*t + phi)
        total_height += eta
        k = newton(lambda k: g*k*tanh(k*h) - (2*pi*f)**2, 0)
        pressure = eta*rho*g*cosh(k*(z + h))/cosh(k*h)
        total_pressure += pressure
    return t, total_height, total_pressure
```

```python
length = 60
sample_frequency = 5

waves = [[.2, 1, pi / 3],
         [.5, .1, pi / 6]]
         
t, y, p = make_waves(length, sample_frequency, waves)
H = ones_like(t) * h
y2 = fft_method(t, p/10000, z, H, 1/sample_frequency, gate=0.1)
plot(t, y)
plot(t, y2)
```
